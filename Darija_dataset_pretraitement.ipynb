**fonction de nettoyage du text**

```python
def nettoyer_text(texte):

    texte = emoji.replace_emoji(texte, replace='')

    # Supprimer diacritiques
    texte = re.sub(r'[\u0610-\u061A\u064B-\u065F]', '', texte)

    texte = re.sub(r'[«»؛:"#$&\'*()+/<=>@\[\]^_`{|}~]', '', texte)

    # Garder lettres arabes + espaces + !?.
    texte = re.sub(r"[^\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF0-9٠-٩\s!%?\.,،]", '', texte)

    texte = re.sub(r'\.$', '', texte)

    texte = texte.replace('\n', ' ')

    # mots_a_supprimer = ['السؤال', 'الإجابة', 'الجواب', 'ج:', 'س:', 'سؤال', 'جاوب:']
    mots_a_supprimer = ['السؤال', 'الإجابة', 'الجواب', 'سؤال', 'جاوب', 'جواب']

    # Supprimer les mots spécifiques
    for mot in mots_a_supprimer:
        texte = re.sub(rf'\b{mot}\b', '', texte)

    # texte = re.sub(r'[جس]\d+', '', texte)
    texte = re.sub(r'\b[سج]\s+', '', texte)

    texte = re.sub(r'[،]+', '،', texte)

    # Nettoyer les espaces
    texte = re.sub(r'\s+', ' ', texte).strip()

    return texte
```

la fonction "extraire_choix" permet de trouver la reponse correct a chaque question de type "multichoice"

```python
import ast
import numpy as np

def extraire_choix(row):
    try:
        # Convertir answer en int s'il est une string
        answer = int(row['answer']) if isinstance(row['answer'], str) else row['answer']

        # Convertir choices en liste si c'est une chaîne
        choices = row['choices']

        if isinstance(choices, (list, np.ndarray)) and 0 <= answer < len(choices):
            return choices[answer]
        else:
            return None
    except Exception as e:
        print(f"Erreur à la ligne {row.name}: {e}")
        return None
```
